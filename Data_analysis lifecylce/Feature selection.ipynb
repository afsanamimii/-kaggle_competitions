Use Lasso for feature selection

```python
from sklearn.linear_model import Lasso,LogisticRegression
from sklearn.feature_selection import SelectFromModel
```

splitting dependent and independent feature

```python
dependent= data['SalePrice']
independent=data.drop(['Id','SalePrice'],axis=1)
```

```python
independent
```

# Applying Lasso for selecting feature

The more the alpha value the less feature will be selected

```python
feature_set=SelectFromModel(LogisticRegression(max_iter=500))
feature_set.fit(independent,dependent)
```

True = The feature is important
False = Not important

```python
feature_set.get_support()
```

Lets check how many features have been selected

```python
total_feature=independent.shape[1]
print("total feature: {}".format(total_feature))

selected_feature=independent.columns[(feature_set.get_support())]
print("selected columns: {}".format(len(selected_feature)))


```

```python
selected_feature
```

```python
dropped_columns= [feature for feature in independent.columns  if feature not  in selected_feature]
```

```python
dropped_columns
```

```python
x=independent[selected_feature]
```

```python
y=dependent
```

```python
x
```

```python
final_data=x.join(y)
final_data
```
